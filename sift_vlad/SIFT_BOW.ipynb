```python
# histogram : list [(q_idx, rank), ...]
def cal_mAP(histogram, topk):
    ap = 0
    print(f"q_num is {len(histogram)}")
    #1개의 쿼리당 AP 계산
    for (q_idx, rank) in histogram:
        if rank+1 <= topk: #rank 0을 기준으로 하므로, +1 하기!
            qap = 1/(rank+1)
            ap += qap
    ap/=len(histogram)
    print(f"mAP@{topk} is {ap}")
    
    return ap
```

```python
import torchvision.transforms.functional as TF

# define a helper function to resize images
def resize_img(image, target_size=(256,256)): # label=(0., 0.), 
    w_orig, h_orig = image.size
    w_target, h_target = target_size
#     cx, cy = label
    image_new = TF.resize(image, target_size)
#     label_new = cx/w_orig*w_target, cy/h_orig*h_target
    return image_new
```

## 2. horizontal flip

```python
def random_hflip(image): #, label
    w, h = image.size
#     x, y = label

    image = TF.hflip(image)
#     label = w-x, y
    return image
```

## 3. vertical flip

```python
def random_vflip(image): #, label
    w, h = image.size
#     x, y = label

    image = TF.vflip(image)
#     label = x, w-y
    return image
```

## 4. shift or translate image

```python
# define a helper function to randomly shift or translate images in either direction
def random_shift(image, max_translate=(0.2, 0.2)):
    w, h = image.size
    max_t_w, max_t_h = max_translate
#     cx, cy = label
    trans_coef = np.random.rand() * 2 - 1
    w_t = int(trans_coef * max_t_w * w)
    h_t = int(trans_coef * max_t_h * h)
    image = TF.affine(image, translate=(w_t, h_t), shear=0, angle=0, scale=1)
#     label = cx + w_t, cy + h_t
    return image
```

## 1. SIFT
    - Lowe 교수
    - keypoint 추출 알고리즘
        1. scale invariant
            : octave(scale이 다른) 별, DOG(Difference of Ggaussian : blur(res이 다른) ) ->  극대, 극소값 찾기 + 나쁜 keypoint 제거(코너, 엣지는 살리고, flat은 빼고)
        2. rotaion invariant
            : 1번에서 찾은 kp의 주변 grid에서 patch에 대해 gradient의 크기/방향 계산 -> 크기: kp 중심으로부터 가우시안 분포 가중치/ 방향: histogram으로 가장 최빈 대표 방향(20~29 : 24.5도)
            : 각각 패치를 또 8개 방향 bin histogram으로 merge하고, 그리고 전체 패치에 대해 grid의 대표 방향으로 빼준다. -> 즉, offset, keypoint 대표 방향이 정해지면 상대적인 방향을 알 수 있다.
        3. illumination invariant
            : pixel 값 normalization
        4. viewpoint invariant
            : 약간 view가 변하는 것에 대해서는 robust

## 2. RootSIFT
    1. Metrics
        - L2(euclidean) < Chi(chi-squared distance)/ Hellinger Kernel
    2. Modify Descriptor(Directly)
        - simple algebraic extension -> it's like the effect of using "Hellinger Kernel" (still L2)
    
        *simple algebraic extension    
            - Step 1: Compute SIFT descriptors using your favorite SIFT library.
            - Step 2: L1-normalize each SIFT vector.
            - Step 3: Take the square root of each element in the SIFT vector. Then the vectors are L2-normalized.
    
    3. Effect
        - dramatically improved result
            - matching keypoints
            - clustering SIFT des
            - quantizing to form of BoW

```python
# import the necessary packages
import numpy as np
import cv2

class RootSIFT:
	def __init__(self, max_kps):
		# initialize the SIFT feature extractor
		self.max_kps = max_kps
		self.extractor = cv2.SIFT_create(self.max_kps)
	def compute(self, image, kps, eps=1e-7):
		# compute SIFT descriptors
		kps, descs = self.extractor.detectAndCompute(image,kps)
		# if there are no keypoints or descriptors, return an empty tuple
		if len(kps) == 0:
			return ([], None)
		# apply the Hellinger kernel by first L1-normalizing and taking the
		# square-root(L2-noramlized)
		descs /= (descs.sum(axis=1, keepdims=True) + eps)
		descs = np.sqrt(descs)
		#descs /= (np.linalg.norm(descs, axis=1, ord=2) + eps)
		# return a tuple of the keypoints and descriptors
		return (kps, descs)
```

### 3.1 ORB feature_extractor

```python
def feature_extractor(file_path):
    '''
    Description: extract feature from given image
    Input: file_path - image path
    Output: des - a list of descriptors of all the keypoint from the image
    '''
    detector = cv2.ORB_create(nfeatures=1000)
    img = cv2.imread(file_path)
    kp = detector.detect(img, None) #find the keypoint
    _, des = detector.compute(img, kp) #compute the descriptor

    return des
```

```python
# wordcnt = 6500 ##hyper-params

def calcFeatVec(features, centers, wordcnt):
	featVec = np.zeros((1, wordcnt)) #voca k개에 대해서 histogram을 만들겠다.
	for i in range(0, features.shape[0]):
		fi = features[i]
        # np.tile des[0] : (1, 128) * k 반복 -> (k, 128)
		diffMat = np.tile(fi, (wordcnt, 1)) - centers #k개의 중심점으로부터 어느 것이 가장 거리가 먼지를 보겠다!
		sqSum = (diffMat**2).sum(axis=1)
		dist = sqSum**0.5
		sortedIndices = dist.argsort()
		idx = sortedIndices[0] # index of the nearest center --> 가장 가까운 중심점에 대해서(아래)
		featVec[0][idx] += 1 #해당 중심점 +1 카운트 한다.
	return featVec.astype('float32') #k개의 histogram을 만들겠다! --> feature vector!
```

```python
# name_list : test_q_name_sorted, test_db_name_sorted
# path_list : test_q_path_sorted, test_db_path_sorted

def quatization(wordcnt, featset):
    
    # # root path
    # q_root_path = str(Path(test_q_path).parent) + '/'
    # db_root_path = str(Path(test_db_path).parent) + '/'
    
    ######DB Quantization
    
    # load learned vocabulary (# 저장폴더 확인!)
    centers = np.load("/home/Signboard_Retrieval/roadview_384_crop/test/checkpoint/voca/" + f"test_db_voca_{wordcnt}_{featset}_rs" + ".npy", \
       allow_pickle=True) #labels, centers  #################################새롬게 배운 코드북으로 변경 60 -> 2500/ rs(RootSIFT) ####################
    # list
    db_featVec_set = []
    
    for idx, db_path in enumerate(test_db_path_sorted): #name --> path
        
        try:
            # check progress
    #         print(f"db {idx+1}/{len(test_db_name_sorted)}")

            # db_path = db_root_path + db_name + '.png' #'.jpg' ###이미지 확장자 주의! --->> test_db_path_sorted : db_path
            img = cv2.imread(db_path)
            # calculate sift feat.
            # des = calcSiftFeature(img)
            des = calcRootSiftFeature(img) #RootSIFT
            # quantize feature Vector by vocabulary of k-means
            featVec = calcFeatVec(des, centers, wordcnt)
            db_featVec_set.append(featVec)
        except:
            print(db_path)
            print("extract_feature_error")
    
    #Reshape
    db_featVec_set = np.array(db_featVec_set).reshape(len(db_featVec_set), -1) #(개수, 차원)
    
    ######Query featVec_Set
    q_featVec_set = []
    
    for idx, q_path in enumerate(test_q_path_sorted): #name --> path
        
        try:
            # check progress
    #         print(f"query {idx+1}/{len(test_q_name_sorted)}")

            # q_path = test_q_root + q_name + '.png' #'.jpg'
            img = cv2.imread(q_path)
            # calculate sift feat.
            # des = calcSiftFeature(img)
            des = calcRootSiftFeature(img) #RootSIFT
            # quantize feature Vector by vocabulary of k-means
            featVec = calcFeatVec(des, centers, wordcnt)
            q_featVec_set.append(featVec)
        except:
            print(q_path)
            print("extract_feature_error")

    #Reshape
    q_featVec_set = np.array(q_featVec_set).reshape(len(q_featVec_set), -1) #(개수, 차원)
    
    return db_featVec_set, q_featVec_set


def retrieval_faiss(db_featVec_set, q_featVec_set):
    
    ######make_db
    #make db (or with custom index)
    #1. L2
    index = faiss.IndexFlatL2(db_featVec_set.shape[1]) # k-mean 차원으로 l2 dist 유사도 구한다!!
    
    #2. Cosine Similiarity
    # faiss.normalize_L2(db_featVec_set)
    # faiss.normalize_L2(q_featVec_set)
    # index = faiss.IndexFlatIP(db_featVec_set.shape[1]) # k-mean 차원으로 Cosine Similiarity로 구한다!!
    
    #Custom Index
#     index = faiss.IndexIDMap2(index)
#     index.add_with_ids(db_featVec_set, db_ids)
    #test
    index.add(np.array(db_featVec_set))
    
    #Check db
    print(f"check db num: {index.ntotal}-{len(test_db_name_sorted)}") ### ex) 개수 확인 - 1352 개의 DB가 모두 잘 들어갔구나!
    
    ######retrieval
    #faiss - retrieval
    distances, indices = index.search(q_featVec_set, index.ntotal) #(query feature set, k) -> topk #error! 100등 안에 것만 보기로 했구나..ㅠ
    
    #make dict
    result_dic = {}
    
    # query 1개의 topk개의 db images
    for idx, _ in enumerate(test_q_path_sorted):
        
        result_dic[idx] = (distances[idx], indices[idx]) #(distance_list of query, index_list of query)
    
    return result_dic, index
```

### - faiss_kmeans

```python
class FaissKMeans:
    def __init__(self, n_clusters=10, n_init=100, max_iter=300):
        self.n_clusters = n_clusters
        self.n_init = n_init
        self.max_iter = max_iter
        self.kmeans = None
        self.cluster_centers_ = None
        self.inertia_ = None

    def fit(self, X, y):
        self.kmeans = faiss.Kmeans(d=X.shape[1],
                                   k=self.n_clusters,
                                   niter=self.max_iter,
                                   nredo=self.n_init)
        self.kmeans.train(X.astype(np.float32))
        self.cluster_centers_ = self.kmeans.centroids
        self.inertia_ = self.kmeans.obj[-1]

    def predict(self, X):
        return self.kmeans.index.search(X.astype(np.float32), 1)[1]
```

### - mAP_code

```python
def cal_mAP(rank_bin_list, top_k):
    rank_bin_arr = np.array(rank_bin_list)   #쿼리에 대한 db의 이미지 유사도 결과를 T/F로 표현하기 위함
    ap = 0.
    print(len(rank_bin_arr))
    #1개의 쿼리당 AP 계산
    for q_idx, item in enumerate(rank_bin_arr[:, :top_k]): #I[: , :top_k] = [0 256 2 4 8] 
        ## 90 PRED = [ True False False False False]\
        qap=.0
        rank=-1        
        for n, v in enumerate(item,start=1):
            if v:
                qap=1/n
                rank=n
                break        
        ap+=qap    
        print(f'[Query {q_idx}] Rank: {rank}, AP: {qap}, mAP: {ap/(q_idx+1)}')
    ap/=len(rank_bin_arr)    
    print(f"mAP is {ap}")

    return ap
```
